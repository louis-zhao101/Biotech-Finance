```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests as requests
import pandas as pd
import numpy as np
```

```python
URL = 'https://old.nasdaq.com/markets/ipos/activity.aspx?tab=pricings&month=2020-03'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find('table')
tablerows = table.find_all('tr')
combined = []
for tr in tablerows:
    td = tr.find_all('td')
    if not td:
        continue
    row = [i.text for i in td]
    combined.append(row)
data = np.array(combined)

dataset = pd.DataFrame({'Company Name': data[:,0], 'TIC' : data[:,1], 'Exchange' : data[:,2],
                       'Price': data[:,3], 'Shares': data[:,4], 'Offer Amount': data[:,5], 'Date': data[:,6] })
dataset
```

```python
from datetime import datetime
baseURL = 'https://old.nasdaq.com/markets/ipos/activity.aspx?tab=pricings&month='
dates = pd.date_range('1980-01', '2020-03', freq = "M")
str_dates = []
for index in range(len(dates)):
    temp=str(dates[index])
    str_dates.append(temp)
    str_dates[index] = str_dates[index][0:7]
```

```python
URLS = []
for val in str_dates:
    tempURL = baseURL + val
    URLS.append(tempURL)
```

```python
month = []
for val in URLS:
    page = requests.get(val)
    soup = BeautifulSoup(page.content, 'html.parser')
    orange = soup.find(class_ = "orange-box")
    if orange !=None:
        continue
    table = soup.find('table')
    tablerows = table.find_all('tr')
    for tr in tablerows:
        td = tr.find_all('td')
        if not td:
            continue
        row = [i.text for i in td]
        month.append(row)
total = np.array(month)

totaldataset = pd.DataFrame({'Company Name': total[:,0], 'TIC' : total[:,1], 'Exchange' : total[:,2],
                      'Price': total[:,3], 'Shares': total[:,4], 'Offer Amount': total[:,5], 'Date': total[:,6] })
```
